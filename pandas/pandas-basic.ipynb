```python
# Merging DataFrames | inner, outer, left, right joins
df1 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df2 = pd.DataFrame({
    'id' : [3,4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})
df2.head()
```

```python
df1.head()
```

```python
inner_result = pd.merge(df1, df2, how='inner', on='id') # merge based on the 'id' column with an inner join
inner_result
```

```python
outer_result = pd.merge(df1, df2, how='outer', on='id')
outer_result
```

```python
left_result = pd.merge(df1, df2, how='left', on='id')
left_result
```

```python
right_result = pd.merge(df1, df2, how='right', on='id')
right_result
```

```python
df3 = pd.DataFrame({
    'id1' : [1,2,3,4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df4 = pd.DataFrame({
    'id2' : [3,4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})

inner_result = pd.merge(df3, df4, how='inner', left_on='id1', right_on='id2')
inner_result
```

```python
# Concatenating DataFrames (combining two dataframes)
df1 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['Keerti', 'Amit', 'Educosys', 'YouTube']
})

df2 = pd.DataFrame({
    'id' : [3,4, 5, 6],
    'course' : ['GenAI', 'CPP', 'HLD', 'LLD']
})

result = pd.concat([df1, df2], axis=0, ignore_index=True) # stacked vertically, ignore_index=True to reset the index
result
```

```python
result2 = pd.concat([df1, df2], axis=1) # stacked horizontally
result2
```

```python
result3 = pd.concat([df1, df2], axis=0, keys=['df1', 'df2']) # stacked vertically with keys to create a hierarchical/ multi level index
result3
```

```python
# GroupBy
"""
groupby is like sorting items into different buckets and then performing calculations on each bucket separately.

Think of it as organizing a classroom of students:


You separate students into groups based on one characteristic (like gender)
Then you calculate something for each group (like average height)
"""
obesity_data.groupby('Gender').count() # group the data based on the Gender column and count the number of occurrences in each group
```

```python
obesity_data.groupby('Gender')['Height'].mean()
"""
This code calculates the mean height for each gender group in the DataFrame.

It works by:


Using groupby('Gender') to split the data into groups based on gender values
Selecting just the 'Height' column from each group with ['Height']
Calculating the average height within each gender group using mean()
The result will be a pandas Series with gender categories as the index and their respective average heights as values. Based on your data sample (which shows only females), it would return a Series with at least one entry for "Female" and its corresponding average height
"""
```

```python
obesity_data.groupby('Gender')['Weight'].median()
```

```python
obesity_data.groupby('Gender')['Weight'].agg(['sum', 'mean']) # calculate the sum and mean of the 'Weight' column for each
```
